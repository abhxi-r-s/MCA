```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.plot(x, y1, label='Prime Numbers')
plt.plot(x, y2, label='Even Numbers')

plt.title('Prime vs Even Numbers')
plt.xlabel('X axis')    
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100) 

fig, axs = plt.subplots(2,2,figsize=(5,4))

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title('Sine Wave')

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title('Cosine Wave')

axs[1,0].plot(x, np.tan(x))
axs[1,0].set_title('Tangent Wave')

axs[1,1].plot(x, np.exp(x))
axs[1,1].set_title('Exponential Wave')

plt.tight_layout()
plt.show()
```

```python
p = ["java", "python", "php", "js", "c#", "c++"]
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# i 

plt.figure(figsize=(5,4))
plt.bar(p,popularity)

plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')

plt.show()

# ii 
plt.figure(figsize=(5,4))
plt.barh(p,popularity,color='red')
plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()


# iii
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']
plt.figure(figsize=(5,4))
plt.bar(p,popularity,color=colors)
plt.title('Popularity of Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
groups = np.arange(1, 6)
men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

# Bar width
width = 0.35

# Plot
plt.figure(figsize=(8, 5))
plt.bar(groups - width/2, men, width=width, label='Men',color='blue')
plt.bar(groups + width/2, women, width=width, label='Women',color='pink')

# Labels
plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by Group and Gender')
plt.xticks(groups)
plt.legend()

plt.show()
```

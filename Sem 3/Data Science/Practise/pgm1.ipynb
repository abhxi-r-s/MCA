```python
import numpy as np
```

```python
print("Multi-D Matrix")
row=int(input("Enter the number of row :"))
col=int(input("Enter the number of column :"))
matrix=[]

for i in range(0,row):
    current_row=[]
    for j in range(0,col):
        values=int(input(f"Enter the number in position {i}{j} :"))
        current_row.append(values)
    matrix.append(current_row)
print(matrix)
    
```

```python
print("MAtrix 2")
row2=int(input("Enter the rows count:"))
col2=int(input("Enter the columns count :"))
matrix2=[]

for i in range(0,row2):
    current_row2=[]
    for j in range(0,col):
        values=int(input(f"Enter the value in the position {i}{j} :"))
        current_row2.append(values)
    matrix2.append(current_row2)

for r in matrix2:
    print(r)
    
```

```python
print("Transpose")
trans=np.transpose(matrix)
print(trans)
```

```python
print("Trace/Sum of main diagonal elements")
tra=np.trace(matrix)
print(tra)
```

```python
print("Determinant")
dett=np.linalg.det(matrix)
print(dett)
```

```python
print("Inverse")
invv=np.linalg.inv(matrix)
print(invv)
```

```python
print("eigen values and vectors")
ei=np.linalg.eig(matrix)
print(ei)
eigg=np.linalg.eigvals(matrix)
print(eigg)
```

```python
print("Dot Product")
if(row==row2 and col==col2):
    lin=np.dot(matrix,matrix2)
    print(lin)
else:
    print("Unable to find")
```

```python
x=np.linspace(0,10,100)
fig,axs=plt.subplots(2,2,figsize=(5,4))

axs[0][0].plot(np.sin(x))
axs[0][0].set_title("Sin wave")

axs[0][1].plot(np.cos(x))
axs[0][1].set_title("Cos wave")

axs[1][0].plot(np.tan(x))
axs[1][0].set_title("Tan wave")

axs[1][1].plot(np.exp(x))
axs[1][1].set_title("Exponential wave")
plt.show()
```
